## 5. FEATURE EXTRACTION

```python
import numpy as np
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.width', 170)

def load_titanic():
    data = pd.read_csv("titanic.csv")
    return data

df = load_titanic()
df.head()
```

### NA_FLAG, BOOL, BINARY

```python
df["NEW_CABIN_BOOL"] = df["Cabin"].isnull().astype('int')
df.head()
```

```python
df["NEW_CABIN_BOOL"] = df["Cabin"].notnull().astype('int')
df.head()
```

```python
df.loc[((df['SibSp'] + df['Parch']) > 0), "NEW_IS_ALONE"] = "NO"
df.loc[((df['SibSp'] + df['Parch']) == 0), "NEW_IS_ALONE"] = "YES"
df.head()
```

### LETTER COUNT

```python
df["NEW_NAME_COUNT"] = df["Name"].str.len()
df.groupby("NEW_CABIN_BOOL").agg({"NEW_NAME_COUNT":"mean"})
```

### WORD COUNT

```python
df["NEW_NAME_WORD_COUNT"] = df["Name"].apply(lambda x: len(str(x).split(" ")))
df.head()
```

### CATCHING SPECIAL PATTERNS

```python
df["NEW_NAME_DR"] = df["Name"].apply(lambda x: len([x for x in x.split() if x.startswith("Dr")]))
df.sort_values(by=["NEW_NAME_DR"], ascending=False).head()
```

```python
df.groupby("NEW_NAME_DR").agg({"Survived": ["mean", "count"]})
```

### EXTRACTING TITLE

```python
df = load_titanic()
df['NEW_TITLE'] = df.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
df.head()
```

```python
df[["NEW_TITLE", "Survived", "Age"]].groupby(["NEW_TITLE"]).agg({"Survived": "mean", 
                                                                 "Age": ["count", "mean"]})
```

### INTERACTIONS

```python
df["NEW_FAMILY_SIZE"] = df["SibSp"] + df["Parch"] + 1

df.loc[(df['Sex'] == 'male') & (df['Age'] <= 21), 'NEW_SEX_CAT'] = 'youngmale'
df.loc[(df['Sex'] == 'male') & ((df['Age'] > 21) & (df['Age']) < 50), 'NEW_SEX_CAT'] = 'maturemale'
df.loc[(df['Sex'] == 'male') & (df['Age'] > 50), 'NEW_SEX_CAT'] = 'seniormale'
df.loc[(df['Sex'] == 'female') & (df['Age'] <= 21), 'NEW_SEX_CAT'] = 'youngfemale'
df.loc[(df['Sex'] == 'female') & ((df['Age'] > 21) & (df['Age']) < 50), 'NEW_SEX_CAT'] = 'maturefemale'
df.loc[(df['Sex'] == 'female') & (df['Age'] > 50), 'NEW_SEX_CAT'] = 'seniorfemale'

df["NEW_AGExPCLASS"] = df["Age"] * df["Pclass"]
df.head()
```

```python
df[["NEW_SEX_CAT", "Survived"]].groupby(["NEW_SEX_CAT"]).agg({"Survived": ["count", "mean"]})
```
